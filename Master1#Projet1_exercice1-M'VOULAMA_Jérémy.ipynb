```python
# Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt

# Définition des paramètres du problème
T = 3          # Temps final
u0 = 1.0       # Condition initiale
N = 24         # Nombre de sous-intervalles
Y = 6          # Autre valeur de N pour comparaison
h = T / N      # Pas de temps pour N
h2 = T / Y     # Pas de temps pour Y

# Définition de la fonction pour la solution exacte
def SolEx(t):
    return u0 * np.exp(-4 * t)

# Création de vecteurs de temps en fonction de N et Y
t = np.linspace(0, T, N + 1)
t2 = np.linspace(0, T, Y + 1)
print(t)
print(t2)

# Définition de la fonction f(t, u) dans l'EDO
def f(t, u):
    return -4 * u

# Implémentation du schéma d'Euler explicite pour N = 24
def EulerExp(t):
    u = np.zeros(len(t))
    u[0] = u0
    for i in range(len(t) - 1):
        u[i + 1] = u[i] + h * f(t[i], u[i])
    return u

# Implémentation du schéma d'Euler explicite pour N = 6
def EulerExp_0(t):
    u = np.zeros(len(t))
    u[0] = u0
    for i in range(len(t) - 1):
        u[i + 1] = u[i] + h2 * f(t[i], u[i])
    return u

# Calcul de la solution exacte, et des solutions approchées avec N = 24 et N = 6
y = SolEx(t)
y1 = EulerExp(t)
y2 = EulerExp_0(t2)

# Tracé des résultats
plt.plot(t, y, "r", label="Solution Exacte")     # Solution exacte en rouge
plt.plot(t, y1, "b-+", label="N = 24")            # Solution avec N = 24 en bleu avec des marqueurs
plt.plot(t2, y2, "k--", label="N = 6")            # Solution avec N = 6 en noir avec un style de ligne en pointillés

# Ajout de titres, légendes et grille
plt.xlabel("t [s]")
plt.ylabel("u(t)")
plt.title("Résolution d'une EDO : schéma d'Euler explicite", fontsize="16")
plt.legend()
plt.grid(color='k')

# Affichage du graphique
plt.show()
```

```python
# Définition de la fonction RKutta pour RK2
def RKutta(t):
    u = np.zeros(len(t))
    u[0] = u0
    for i in range(len(t) - 1):
        k1 = h * f(t[i], u[i])
        k2 = h * f(t[i] + h, u[i] + k1)
        u[i + 1] = u[i] + 1/2 * (k1 + k2)
    return u

# Définition de la fonction RKutta_2 pour RK4
def RKutta_2(t):
    u = np.zeros(len(t))
    u[0] = u0
    for i in range(len(t) - 1):
        k1 = h * f(t[i], u[i])
        k2 = h * f(t[i] + h2, u[i] + k1/2)
        k3 = h * f(t[i] + h2, u[i] + k2/2)
        k4 = h * f(t[i] + h, u[i] + k3/2)
        u[i + 1] = u[i] + 1/6 * (k1 + 2*k2 + 2*k3 + k4)
    return u

# Calcul de la solution exacte, et des solutions approchées avec RK2 et RK4
y = SolEx(t)
y1 = RKutta(t)
y2 = RKutta_2(t2)

# Tracé des résultats
plt.plot(t, y, "b-", label="Solution Exacte")       # Solution exacte en bleu avec un trait plein
plt.plot(t, y1, "g-*", label="RK2 (N = 24)")      # Solution avec RK2 et N = 24 en vert avec des étoiles
plt.plot(t2, y2, "r--", label="RK4 (N = 6)")      # Solution avec RK4 et N = 6 en rouge avec un style de ligne en pointillés

# Ajout de titres, légendes et grille
plt.xlabel("t [s]")
plt.ylabel("u(t)")
plt.title("Résolution d'une EDO : schéma de Runge-Kutta 4", fontsize="16")
plt.legend()
plt.grid(color='k')

# Affichage du graphique
plt.show()
```
